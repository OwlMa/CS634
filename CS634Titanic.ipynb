```python
import os
import tarfile
import pandas as pd
import numpy as np
from pandas import Series,DataFrame

#input the data first

TITANIC_PATH="e:\\Midtermproject of DataMining\\Project2"
trainname="train.csv"
testname="test.csv"

def load_titanic_data(titanic_path=TITANIC_PATH,dataname=trainname):
    csv_path=os.path.join(titanic_path,dataname)
    return pd.read_csv(csv_path)

titanic_train=load_titanic_data()
titanic_train.head()
titanic_test=load_titanic_data(TITANIC_PATH,testname)
titanic_test.head()
```

```python
full=titanic_train.append(titanic_test,ignore_index=True)
full.shape
```

```python
full.info()
# we can see the missing data in the train set
# Age, Cabin and Embarked are three missing data, so we need to fill these three kinds of data
```

```python
full['Age']=full['Age'].fillna(full['Age'].mean())
full=full.drop(["Cabin"],axis=1)
full=full.drop(['Name'],axis=1)
full['Embarked']=full['Embarked'].fillna('S')
full['Fare']=full['Fare'].fillna(full['Fare'].mean())
full.info()
```

```python
#change the text feature to numerical feature 
full['Sex'].value_counts()
sexDict={'male':1,'female':0}
full['Sex']=full['Sex'].map(sexDict)
full.head()
```

```python
embarkedDF=pd.DataFrame()
embarkedDF=pd.get_dummies(full['Embarked'],prefix='Embarked')
embarkedDF.head()
```

```python
full=pd.concat([full,embarkedDF],axis=1)
full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
pclassDF=pd.DataFrame()
pclassDF=pd.get_dummies(full['Pclass'],prefix='Pclass')
pclassDF.head()
```

```python
full=pd.concat([full,pclassDF],axis=1)
full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
from sklearn.preprocessing import scale
age_scaler=scale(full['Age'])
full['Age']=age_scaler
full.head()
```

```python
fare_scaler=scale(full['Fare'])
full['Fare']=fare_scaler
full.head()
```

```python
#corralation
corrDF=full.corr()
corrDF['Survived'].sort_values(ascending=False)
```

```python
full_x=pd.concat([pclassDF,embarkedDF,full['Age'],full['Fare'],full['Sex'],full['SibSp'],full['Parch']],axis=1)
full_x.head()
```

```python
sourceRow = 891
source_X = full_x.loc[0:sourceRow-1,:] 
source_y = full.loc[0:sourceRow-1,'Survived']
pred_X = full_x.loc[sourceRow:,:]
pred_X
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(source_X,source_y,train_size=0.8)
source_y.head()
```

```python
#from sklearn.linear_model import LogisticRegression
#model = LogisticRegression()
#model.fit(train_X,train_y)
#model.score(test_X,test_y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
modelGD = GradientBoostingClassifier()
modelGD.fit(train_X,train_y)
modelGD.score(test_X,test_y)
```

```python
pred_y = modelGD.predict(pred_X)
pred_y = pred_y.astype(int)
passenger_id = full.loc[sourceRow:,'PassengerId']
predDf = pd.DataFrame({'PassengerId':passenger_id, 'Survived':pred_y})
predDf.shape
predDf
```
